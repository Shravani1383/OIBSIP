Encoding labels into Numerical format using label Encoder

```python
x=iris.drop('Species',axis=1)
y=iris['Species'].copy()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y=le.fit_transform(y)
y
```

Spliting of data into Training data and test data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
```

Model Fit Decision Tree classifier

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
dt_model=DecisionTreeClassifier()
dt_model.fit(X_train,y_train)
dt_pred=dt_model.predict(X_test)
print('Accuracy:',metrics.accuracy_score(dt_pred,y_test))
```

Cross Validation

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(dt_model,X_train, y_train, scoring='neg_mean_squared_error', cv=10)
rmse_scores = np.sqrt(-scores)
rmse_scores
```

Confusion Matrix for decision Tree

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
predictions = dt_model.predict(X_test)
print(classification_report(y_test,predictions))
print("Confusion Matrix\n",confusion_matrix(y_test,predictions))
```

Logistic regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
lr_model=LogisticRegression()
lr_model.fit(X_train,y_train)
lr_pred=lr_model.predict(X_test)
print('Accuracy:',metrics.accuracy_score(lr_pred,y_test))
```

Cross Validation

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(lr_model,X_train, y_train, scoring='neg_mean_squared_error', cv=10)
rmse_scores = np.sqrt(-scores)
rmse_scores
```

Confusion Matrix for Logistic Regrssion

```python
predictions = lr_model.predict(X_test)
print(classification_report(y_test,predictions))
print("Confusion Matrix\n",confusion_matrix(y_test,predictions))
```

KNeighbours Classifier

```python
from sklearn.neighbors import KNeighborsClassifier

knn_model = KNeighborsClassifier(n_neighbors=10)
knn_model.fit(X_train, y_train)
pred_knn = knn_model.predict(X_test)
accuracy_knn =print('Accuracy:',metrics.accuracy_score(pred_knn,y_test))
accuracy_knn
```

Cross Validation for knn

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(knn_model,X_train, y_train, scoring='neg_mean_squared_error', cv=10)
rmse_scores = np.sqrt(-scores)
rmse_scores
```

Confusion Matrix for knn
